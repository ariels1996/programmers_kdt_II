## I. Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리 

cf) 라이브러리 vs 프레임워크 : 내용물이 주어져 그것을 활용 vs 틀이 짜여져있어 내용을 채워야 함

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## II. Matplotlib Case Study with Arguments

```python
# 꺾은 선 그래프
plt.plot([1, 2, 3, 4, 5]) # 실제 plotting을 하는 함수 #index가 x, 값들이 y
plt.show() # plt를 확인하는 방법
```

### Figsize : Figure의 크기 조절

```python
plt.figure(figsize=(3, 3)) # plotting을 할 도면을 선언 

plt.plot([1, 2, 3, 4, 5])
plt.show()
```

### 2차함수 그래프 with plot()

```python
x = np.array([1, 2, 3, 4, 5]) # 정의역
y = np.array([1, 4, 9, 16, 25]) # f(x)

plt.plot(x, y)
plt.show()
```

```python
# np.arange(a, b, c) -> c:0.01 점이 찍히는 간격을 정수가 아닌 실수로 정할 수 있다. 

x = np.arange(-10,10,0.01)
plt.plot(x, x**2)

plt.show()
```

```python
# x, y축에 설명 추가하기
x = np.arange(-10,10,0.01)

###
plt.xlabel("x value")
plt.ylabel("f(x) value")
###

plt.plot(x, x**2)

plt.show()
```

```python
# x, y축의 범위를 설정하기

x = np.arange(-10,10,0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

###
plt.axis([-5, 5, 0, 50]) # [x_min, x_max, y_min, y_max]
###

plt.plot(x, x**2)

plt.show()
```

```python
# x, y축의 눈금 설정하기

x = np.arange(-10,10,0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) 

###
plt.xticks([i for i in range(-5, 5, 1)]) # x축의 눈금 설정 -5, -4, -3, ...
plt.yticks([i for i in range(0, 27, 3)]) # y축의 눈금 설정
###

plt.plot(x, x**2)

plt.show()
```

```python
# 그래프에 타이틀 달기

x = np.arange(-10,10,0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) 
plt.xticks([i for i in range(-5, 6, 1)]) # x축의 눈금 설정 -5, -4, -3, ...
plt.yticks([i for i in range(0, 28, 3)]) # y축의 눈금 설정
plt.title("y = x^2 graph")

plt.plot(x, x**2, label="trend")

### legend는 항상 그래프가 그려진 이후에 작성해야한다. 
plt.legend()
###

plt.show()
```

## III. Matplotlib Case Study

### 꺾은선 그래프 (Plot)

시계열 데이터를 나타낼 때 주로 사용합니다. 

```python
x = np.arange(20) # 0~19
y = np.random.randint(0, 20, 20) # 0~19사이에 난수를 20개 생성

# y축을 20까지 5단위로 보이게 만들기
plt.axis([0, 20, 0, 20]) 
plt.yticks([i for i in range(0, 21, 5)]) # y축의 눈금 설정
# plt.yticks([0, 5, 10, 15, 20]) 도 가능

plt.plot(x, y)

plt.show()
```

### 산점도 (Scatter Plot)

상관관계가 없는 데이터들을 나타낼 때 사용합니다. 

```python
plt.scatter(x, y)
plt.show()
```

### 박스 그림 (Box Plot)

분포를 확인하고 싶을 때 사용
- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, Min, Max)

```python
plt.boxplot((x, y))

plt.title("Box plot of x, y")

plt.show()
```

### 막대 그래프 (Bar Plot)

- 범주형 데이터의 '값'과 그 값의 크기를 직사각형으로 나타낸 그림

```python
plt.bar(x, y)
plt.xticks([i for i in range(0, 20, 1)]) # x축의 눈금 설정

plt.show()
```

히스토그램

```python
# cf) Histogram
# 도수 분포를 막대형태로 나타내는 건 비슷하나
# 계급을 나눠서 그룹화한 뒤 표시한다. 

plt.hist(y, bins=np.arange(0, 20, 2))
plt.xticks(np.arange(0, 20, 2)) # x축의 눈금 설정

plt.show()

 
```

### 원형 그래프(Pie Chart)

- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프 
- 다른 그래프에 비해서 **비율** 확인에 용이
- `.pie()`

```python
z = [100, 300, 200, 400]

plt.pie(z, labels=['one', 'two', 'three', 'four'])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

### Matplotlib 를 기반으로 더 다양한 시각화 방법을 제공하는 라이브러리

- 커널밀도그림
- 카운트그림

- 스트립그림
- 히트맵

### Seaborn Import하기

```python
import seaborn as sns
```

### 커널밀도그림 (Kernel Density Plot)

- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린그림
- sns.kdeplot()

```python
# in histogram

x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins=x)
plt.show()
```

```python
# kdeplot
print(y)
sns.kdeplot(y, shade=True)
plt.show()
```

### 카운트그림 (Count Plot)

- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과 
- sns.countplot()

```python
vote_df = pd.DataFrame({"name":['Andy','Bob','Cat'], "vote":[True, False, True]})

vote_df
```

```python
vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x=[False, True], height=vote_count['name'])
plt.show()
```

```python
# sns의 countplot

sns.countplot(x=vote_df['vote'])
plt.show()
```

### 캣 그림 (Cat Plot)

- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- sns.catplot()

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.headad(5)
```

```python
s = sns.catplot(x='WHO Region', y='Confirmed', data=covid, kind='strip')
s.fig.set_size_inches(10, 6)
plt.show()
```

```python
s = sns.catplot(x='WHO Region', y='Confirmed', data=covid, kind='violin')
s.fig.set_size_inches(10, 6)
plt.show()
```

### 스트립그림 (Strip Plot)

- scatter plot 과 유사하게 데이터의 수치를 표현하는 그래프
- sns.stripplot()

```python
sns.stripplot(x='WHO Region', y='Recovered', data=covid)
plt.show()
```

```python
#cf) swarmplot
sns.swarmplot(x='WHO Region', y='Recovered', data=covid)
plt.show()
```

### 히트맵(Heat map)

- 데이터의 행렬을 색상으로 표현해주는 그래프
- sns.heatmap()

```python
# 히트맵 예제
covid.corr()
```

```python
sns.heatmap(covid.corr())

plt.show()
```
